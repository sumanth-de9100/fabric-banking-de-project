```python
from pyspark.sql import functions as F
```

```python
transactions_path="Files/source_data/transactions"

df_txn_raw=(
    spark.read
    .option("header","true")
    .option("infer schema","true")
    .csv(transactions_path)
)
```

```python
display(df_txn_raw)
```

```python
df_txn_bronze=(
    df_txn_raw
    .withColumn("ingestion_ts",F.current_timestamp())
    .withColumn("source_file",F.input_file_name())
)
```

```python
display(df_txn_bronze)
```

```python
(
    df_txn_bronze
    .write
    .mode("append")
    .format("delta")
    .saveAsTable("bronze_transactions")
)
```

```python
customers_path="Files/source_data/customer/customers.json"

df_cust_raw =(
  spark.read
    .option("multiline","true")
    .json(customers_path)
)
```

```python
display(df_cust_raw)
```

```python
df_cust_bronze=( 
    df_cust_raw
    .withColumn("ingestion_ts",F.current_timestamp())
)
```

```python
(
    df_cust_bronze
    .write
    .mode("append")
    .format("delta")
    .saveAsTable("bronze_raw_customers")
)
```

```python
merchant_category="Files/source_data/reference/merchant_risk.csv"

df_merchant_cat=(
    spark.read
    .option("header","true")
    .option("infer schema","true")
    .csv(merchant_category)
)
```

```python
display(df_merchant_cat)
```

```python
(
    df_merchant_cat
    .write
    .mode("append")
    .format("delta")
    .saveAsTable("bronze_merchant_risk")
)
```
